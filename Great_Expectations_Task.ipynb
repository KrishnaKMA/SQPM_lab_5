##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context(mode="ephemeral")
datasource = context.data_sources.add_pandas("pandas_datasource")
data_asset = datasource.add_dataframe_asset("labels_asset")
batch_definition = data_asset.add_batch_definition_whole_dataframe("batch_definition")
```

##6. Define and Create a Data Batch

```python
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

### Expectation 1

```python
# Expectation 1: Pedestrian bounding box X coordinates (top-left) must fall within valid image width bounds (0 to 900 pixels)
expectation1 = gx.expectations.ExpectColumnValuesToBeBetween(
    column="pedestrianLocationX_TopLeft", min_value=0, max_value=900
)
```

### Validate Data Against Expectation 1

```python
result1 = batch.validate(expectation1)
print(result1)
```

### Expectation 2

```python
# Expectation 2: The Timestamp column must have no null/missing values
expectation2 = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="Timestamp"
)
```

### Validate Data Against Expectation 2

```python
result2 = batch.validate(expectation2)
print(result2)
```

### Expectation 3

```python
# Expectation 3: Ground truth image filenames must follow the expected naming convention (e.g., C_001.png)
expectation3 = gx.expectations.ExpectColumnValuesToMatchRegex(
    column="Ground_Truth_View", regex=r"C_\d{3}\.png"
)
```

### Validate Data Against Expectation 3

```python
result3 = batch.validate(expectation3)
print(result3)
```
